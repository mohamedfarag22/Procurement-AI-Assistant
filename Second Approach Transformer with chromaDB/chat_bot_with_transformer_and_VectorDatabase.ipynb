```python
import chromadb
from sentence_transformers import SentenceTransformer
import json

# Initialize model and ChromaDB
client = chromadb.PersistentClient(path="./product_db")
collection = client.get_collection("products")

def retrieve_top_10(collection, query):
    """
    Retrieve and print the top 10 most similar records for a query.
    Properly embeds the query before searching.
    """
    # Step 1: Embed the query using the same model
    query_embedding = model.encode(query).tolist()
    
    # Step 2: Search with the embedded query
    results = collection.query(
        query_embeddings=[query_embedding],  # Use the pre-computed embedding
        n_results=100,
        include=["documents", "distances"]  # IDs are always included
    )
    
    print(f"\nTop 10 results for query: '{query}'")
    print("=" * 80)
    
    for i, (id_, doc, distance) in enumerate(zip(
        results["ids"][0],
        results["documents"][0],
        results["distances"][0]
    )):
        similarity_score = 1 - distance  # Convert distance to similarity (0-1)
        
        print(f"\nResult #{i+1} | ID: {id_} | Similarity: {similarity_score:.3f}")
        print("-" * 50)
        
        try:
            data = json.loads(doc)
            print(json.dumps({
                "Item": data.get("Item Name"),
                "Description": data.get("Item Description"),
                "Total Price" : data.get("Total Price"),
                "Price": data.get("Total Price"),
                "Department": data.get("Department Name"),
                "Date": data.get("Purchase Date")
            }, indent=2))
        except json.JSONDecodeError:
            print(doc[:200] + ("..." if len(doc) > 200 else ""))
    
    print("\n" + "=" * 80)

# Example usage
```

```python
def retrieve_highest_priced(collection, query, pool_size=50, top_k=10):
    """
    1. Finds semantically related purchases
    2. Returns those with highest Total Price values
    """
    # 1. Get initial larger pool of similar records
    results = collection.query(
        query_texts=[query],
        n_results=pool_size,
        include=["documents", "distances"]
    )
    
    # 2. Extract and parse prices
    priced_records = []
    for doc, dist, id_ in zip(results["documents"][0], 
                            results["distances"][0],
                            results["ids"][0]):
        try:
            data = json.loads(doc)
            price = float(data.get("Total Price", 0))
            priced_records.append({
                "id": id_,
                "price": price,
                "similarity": 1 - dist,
                "data": data
            })
        except (json.JSONDecodeError, ValueError):
            continue
    
    # 3. Sort by price (descending) and take top_k
    highest_priced = sorted(priced_records, key=lambda x: x["price"], reverse=True)[:top_k]
    
    # 4. Display results
    print(f"\nTop {top_k} Highest Priced Purchases (from {len(priced_records)} candidates)")
    print("=" * 80)
    for i, record in enumerate(highest_priced):
        print(f"\n#{i+1} | ID: {record['id']}")
        print(f"Price: ${record['price']:,.2f} | Similarity: {record['similarity']:.3f}")
        print("-" * 60)
        print(f"Item: {record['data'].get('Item Name')}")
        print(f"Item: {record['data'].get('Total Price')}")

        print(f"Dept: {record['data'].get('Department Name')}")
        print(f"Date: {record['data'].get('Purchase Date')}")
    
    print("=" * 80)
    return highest_priced
```

```python
retrieve_highest_priced(collection, query="what is the total number of orders for a 2013??", pool_size=50, top_k=10)
```
